```python
# Program to take Name & Marks of student as a input from user and display the bar chart for all.

import pandas as pd
import matplotlib.pyplot as plt
empty=[]
num=[]
while True:
    Name=input("Enter the Student Name:").title()
    Marks=int(input("Enter the Student Marks:"))
    empty.append(Name)
    num.append(Marks)
    choice=input("Do you want to add more student: (Y/N)").title()
    if choice=="Y":
        print(empty)
        print(num)
    else:
        option=input("Do you want to display a chart: (Y/N)").title()
        if option=="Y":
            plt.bar(empty,num,linewidth=5,edgecolor="black")
            plt.title("Student Marks Comparison",fontsize=16)
            plt.xlabel("Student Name",fontsize=16,color="red")
            plt.ylabel("Student Marks",fontsize=16,color="purple")
            plt.show()
            break
        else:
            break
    
```
