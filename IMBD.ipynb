```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("movies.csv")
df.head()
```

```python
# 1.plot a graph to show the top 5 movies based on number of votes.

df1=df.sort_values(by="imbd_votes",ascending=True)
df1=df.head(6)
gb=df1.groupby("title").agg({"imbd_votes":"sum"})
gb=gb.sort_values(by="imbd_votes")
plt.bar(gb.index,gb["imbd_votes"])
plt.xticks(rotation=90)
plt.show()
```

```python
# 2.find out which movie has the maximum number of votes and which genre it belongs to and its duration.
max=df["imbd_votes"].idxmax()
df.loc[max]
```

```python
# 3.find out which movie has the minimum number of votes and which genre it belongs to and its duration.
min=df["imbd_votes"].idxmin()
df.loc[min]
```

```python
# 4.find out movies of each genre which has maximum number of votes.
gn=df.groupby(["genre","title"]).agg({"imbd_votes":"max"})
gn
```
