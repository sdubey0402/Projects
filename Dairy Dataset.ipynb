```python
import pandas as pd
df=pd.read_csv("dairy_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 1. does number of cows in a farm has any dependency on area aquired.

data=df.groupby("Farm Size").agg({"Number of Cows":"count","Total Land Area (acres)":"mean"})
data
```

```python
# 2.what is the average expiry days of each product.

data=df.groupby(["Product Name","Storage Condition"]).agg({"Shelf Life (days)":"mean"})
data=data.sort_values(by="Shelf Life (days)")
data
```

```python
# 3. mention the name of the products with max and min shelf life

data1=df["Shelf Life (days)"].idxmax()
df.loc[data1]
```

```python
data2=df["Shelf Life (days)"].idxmin()
df.loc[data2]
```

```python
# 4. summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them.
data3=df.groupby("Brand").agg({"Total Land Area (acres)":"sum","Number of Cows":"sum","Approx. Total Revenue(INR)":"sum"})
data3

```

```python
# 5. What is the average shelf life for different storage conditions.

data4=df.groupby("Storage Condition").agg({"Shelf Life (days)":"mean"})
data4=data4.sort_values(by="Shelf Life (days)")
data4
```

```python
# my analysis 
# What is total revenue sales channel for brand wise.

data5=df.groupby(["Sales Channel","Brand"]).agg({"Approx. Total Revenue(INR)":"sum"})
data5=data5.sort_values(by="Approx. Total Revenue(INR)")
data5
```
